```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = 'price_stock.csv'
dataframe = pd.read_csv(path)
```

```python
df_transposed = dataframe.T
df_transposed
```

```python
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
```

```python
X = TimeSeriesScalerMeanVariance().fit_transform(df_transposed.iloc[1:])
X.shape
```

```python
X = X.reshape(60,242)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
from tslearn.clustering import TimeSeriesKMeans
```

```python
sz = X.shape[0]
```

```python
distortions = []
K = range(1, 30)
 
for k in K:
    km_sdtw = TimeSeriesKMeans(n_clusters=k, metric="softdtw", max_iter=5,max_iter_barycenter=5,metric_params={"gamma": .5},random_state=0).fit(X)
    distortions.append(km_sdtw.inertia_)
```

```python
plt.figure(figsize=(20,10))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
